# man

```python
class Man:

    def __init__(self,name):
        self.name = name
        print('Initialized!')

    def hello(self):
        print("Hello " + self.name + "!")
    
    def goodbye(self):
        print('Good-bye ' + self.name + "!")
```

```python
m = Man('David')
m.hello()
m.goodbye()
```

# sin_cos_graph

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
