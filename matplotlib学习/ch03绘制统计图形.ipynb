```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 3.1柱状图

```python
x = np.arange(1,6)
y = np.random.randint(1,11,5)
```

```python
x,y
```

```python
plt.figure(figsize=(14,7))
plt.bar(x,y,align='center',color='r',
       tick_label=['A','B','C','D','E'],alpha=.6)
plt.xlabel('test')
plt.ylabel('number')
plt.grid(True,axis='y',ls=':',color='r',alpha=.3)
plt.show()
```

## 3.2条形图

```python
plt.figure(figsize=(14,7))
plt.barh(x,y,align='center',color='r',
       tick_label=['A','B','C','D','E'],alpha=.6)
plt.ylabel('test')
plt.xlabel('number')
plt.grid(True,axis='x',ls=':',color='r',alpha=.3)
plt.show()
```

## 3.3堆积图

```python
y1 = np.random.randint(2,9,5)
```

```python
y1
```

```python
plt.figure(figsize=(14,7))
plt.bar(x,y,align='center',color='r',
        tick_label=['A','B','C','D','E'],label='Class A')
plt.bar(x,y1,align='center',color='b',bottom=y,label='Class B')
plt.legend()
plt.show()
```

### 3.3.2堆积条形图

```python
plt.figure(figsize=(14,7))
plt.barh(x,y,align='center',color='r',
         tick_label=['A','B','C','D','E'],label='Class A')
plt.barh(x,y1,align='center',left=y,color='g',label='Class B')
plt.legend()
plt.show()
```

### 3.4.1多数据并列柱状图

```python
x = np.arange(5)
y = np.random.randint(1,11,5)
y1 = np.random.randint(2,9,5)
```

```python
x,y,y1
```

```python
bar_width = .35
tick_label = ['A','B','C','D','E']
#plt.figure(figsize=(14,7))
plt.bar(x,y,bar_width,color='c',align='center',alpha=.5,label='Class A')
plt.bar(x+bar_width,y1,bar_width,color='b',
        align='center',label='Class B',alpha=.5)
plt.xticks(x+bar_width/2,tick_label)
plt.legend()
plt.show()
```

### 3.4.2多数据平行条形图

```python
plt.barh(x,y,bar_width,
         color='c',align='center',
         label='Class A',hatch='///')
plt.barh(x+bar_width,y1,
         bar_width,color='r',
         align='center',label='Class B',hatch='+')
plt.yticks(x+bar_width/2,tick_label)
plt.legend()
plt.show()
```

## 3.6堆积折线图、间断条形图、阶梯图

### 3.6.1stackplot()堆积折线图

```python
x = np.arange(1,6,)
```

```python
y = np.random.randint(0,7,5)
y1 = np.random.randint(1,8,5)
y2 = np.random.randint(1,7,5)
y,y1,y2
```

```python
labels = ['A','B','C']
colors = ['r','g','b']
plt.stackplot(x,y,y1,y2,labels=labels,colors=colors)
plt.legend()
plt.show()
```

### 3.6.2 broken_barh()绘制间断条形图

```python
plt.broken_barh([(30,100),(180,50),(260,70)],(20,8),facecolors='r')
plt.broken_barh([(60,90),(190,20),(230,30),(280,60)],(10,8),facecolors=('r','g','b','c'))
plt.xlim(0,360)
plt.ylim(5,35)
plt.xticks(np.arange(0,361,60))
plt.yticks([15,25],['Opera A','Opera B'])
plt.grid(ls='-',lw=1,color='gray')
plt.title('Time diff')
plt.show()
```

### 3.6.3step()绘制阶梯图

```python
x = np.linspace(1,10,10)
y = np.sin(x)
x,y
```

```python
plt.step(x,y,'c',where='pre',lw=2)#pre代表z
plt.show()
```

```python
plt.step(x,y,'c',where='post',lw=2)
plt.show()
```
