### 1.3.5grid()绘制刻度线的网格线

```python
x = np.linspace(0.05,10,1000)
y = np.sin(x)
```

```python
plt.figure(figsize=(14,7))
plt.plot(x,y,ls='-',lw=3,label='plot figure')
plt.legend()
plt.xlim(0,10)
plt.ylim(-2.0,3.0)
plt.grid(ls='-',c='r')
plt.show()
```

### 1.3.6axhline()绘制平行于x轴的水平参考线

```python
plt.figure(figsize=(14,7))
plt.plot(x,y,c='r',ls='-',lw=3,label='plot figure')
plt.legend()
plt.axhline(y=4,ls='--',lw=3,c='b')#水平线
plt.xlabel('$y=sin(x)$')
plt.axvline(x=3,ls='-',lw=3,c='purple')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

### 1.3.7axvspan()——绘制垂直于x轴的参考区域

```python
plt.figure(figsize=(14,7))
plt.plot(x,y,c='r',ls='-',lw=3,label='plot figure')
plt.legend()
plt.xlim(0,10)
plt.ylim(-1,1)
plt.xlabel('$y=sin(x)$')
plt.axvspan(4,6,facecolor='y',alpha=0.3)
plt.axhspan(0,0.5,facecolor='y',alpha=0.3)
plt.show()
```

### 1.3.8annotate()——添加图形内容细节的指向性注释文本

```python
plt.figure(figsize=(14,7))
plt.plot(x,y,ls='-',lw=3,c='c',label='plot figure',)
plt.legend(fontsize=24,loc='lower right')
plt.annotate('maximum',xy=(np.pi/2,1.0),xytext=((np.pi/2)+1.0,.8),weight='bold',c='b',arrowprops=dict(arrowstyle='->',connectionstyle='arc3',color='purple'))
plt.show()
```

### 1.3.9text()——添加图形内容细节的无指向型注释文本

```python
plt.figure(figsize=(14,7))
plt.plot(x,y,c='c',lw=3,ls='-',label='plot figure')
plt.legend(fontsize=30,loc='lower right')
plt.text(3.10,0.09,'y=sin(x)',weight='bold',c='b',fontsize=30)
plt.show()
```

### 1.3.10title()——添加图形内容的标题

```python
plt.figure(figsize=(14,7))
plt.plot(x,y)
plt.title('y=sin(x)',fontsize=34)
plt.show()
```

# 综合运用所有内容

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
from matplotlib import cm as cm
```

```python
x = np.linspace(0.5,3.5,100)
y = np.sin(x)
y1 = np.random.randn(100)
plt.figure(figsize=(14,7))
plt.scatter(x,y1,c='0.25',label='scatter figure')
plt.plot(x,y,ls='--',lw=2,label='plot figure')
for spine in plt.gca().spines.keys():
    if spine == 'top' or spine == 'right':
        plt.gca().spines[spine].set_color('none')
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().yaxis.set_ticks_position('left')
plt.xlim(0.0,4.0)
plt.ylim(-3.0,3.0)
plt.ylabel('y_axis')
plt.xlabel('x_axis')
plt.grid(True,ls=':',lw=2)
plt.axhline(y=1.0,c='r',ls='--',lw=2)
plt.axvspan(1.0,2.0,facecolor='y',alpha=0.3)
#plt.annotate('maximum',xy=(np.pi/2,1.0),xytext=((np.pi/2)+0.15),1.5),weight='bold',c='r',arrowprops=dict(arrowstyle='->',connectionstyle='arc3',c='r')
plt.annotate('spines',xy=(0.75,-3),xytext=(0.35,-2.25),weight='bold',c='b',arrowprops=dict(arrowstyle='->',connectionstyle='arc3',c='b'))
plt.annotate('',xy=(0,-2.78),xytext=(0.4,-2.32),arrowprops=dict(arrowstyle='->',connectionstyle='arc3',c='b'))
plt.annotate('',xy=(3.5,-2.98),xytext=(3.6,-2.70),arrowprops=dict(arrowstyle='->',connectionstyle='arc3',c='b'))
plt.text(3.6,-2.70,"'|'is tickline",weight='bold',c='b')
plt.text(3.6,-2.95,"3.5 is tickline",weight='bold',c='b')
plt.title('structure of matplotlib')
plt.legend()
plt.show()
```
