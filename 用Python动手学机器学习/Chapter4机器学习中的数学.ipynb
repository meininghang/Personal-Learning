### 4.5.3绘制梯度的图形

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#定义函数和求导公式
def f(w0,w1):
    return w0**2 + 2*w0*w1 + 3

def df_dw0(w0,w1):
    return 2*w0 +2*w1

def df_dw1(w0,w1):
    return 2*w0
```

```python
w_range = 2
dw = 0.25
w0 = np.arange(-w_range,w_range+dw, dw)
w1 = np.arange(-w_range,w_range+dw, dw)
```

```python
ww0, ww1 = np.meshgrid(w0, w1)
```

```python
ff = np.zeros((len(w0),len(w1)))
dff_dw0 = np.zeros((len(w0),len(w1)))
dff_dw1 = np.zeros((len(w0),len(w1)))
for i0 in range(len(w0)):
    for i1 in range(len(w1)):
        ff[i1, i0] = f(w0[i0],w1[i1])
        dff_dw0[i1,i0] = df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0] = df_dw1(w0[i0],w1[i1])
```

```python
plt.figure(figsize=(9,4))
plt.subplots_adjust(wspace=0.3)
plt.subplot(1,2,1)
cont = plt.contour(ww0,ww1,ff,10,colors='blue')#显示等高线
cont.clabel(fmt='%d',fontsize=8)
plt.xticks(range(-w_range,w_range+1,1))
plt.yticks(range(-w_range,w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.xlabel('$w_0$',fontsize=14)
plt.ylabel('$w_1$',fontsize=14)

plt.subplot(1,2,2)
plt.quiver(ww0,ww1,dff_dw0,dff_dw1)#显示f的梯度向量
plt.xlabel('$w_0$',fontsize=14)
plt.ylabel('$w_1$',fontsize=14)
plt.xticks(range(-w_range,w_range+1,1))
plt.yticks(range(-w_range,w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.show()
```

### 4.7.6Softmax函数

```python
def softmax(x0,x1,x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0)/u, np.exp(x1)/u, np.exp(x2)/u
```

```python
y = softmax(2,1,-1)
```

```python
y
```

```python
np.round(y,2)
```

```python
np.sum(y)
```

#### 绘制Softmax函数图像

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
xn = 20
x0 = np.linspace(-4,4,xn)
x1 = np.linspace(-4,4,xn)
```

```python
y = np.zeros((xn,xn,3))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0, :] = softmax(x0[i0],x1[i1],1)
```

```python
xx0,xx1 = np.meshgrid(x0,x1)
plt.figure(figsize=(14,7))
for i in range(2):
    ax = plt.subplot(1,2,i+1,projection='3d')
    ax.plot_surface(xx0,xx1,y[:,:,1],rstride=1,cstride=1,alpha=0.3,color='blue',edgecolor='red')
    ax.set_xlabel('$x_0$',fontsize=14)
    ax.set_ylabel('$x_1$',fontsize=14)
    ax.view_init(40,-125)
plt.show()
```

### 4.7.8 高斯函数

```python
def gauss(mu,sigma,a):
    return a * np.exp(-(x-mu)**2 / (2*sigma**2))
```

### 4.7.9 二维高斯函数

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#高斯函数
def gauss(x,mu,sigma):#x是Nx2矩阵，mu是模为2的向量，sigma是2x2矩阵
    N,D = x.shape
    c1 = 1 / (2*np.pi)**(D/2)
    c2 = 1 / (np.linalg.det(sigma)**(1/2))
    inv_sigma = np.linalg.inv(sigma)
    c3 = x - mu
    c4 = np.dot(c3,inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5 + c4[:,d]*c3[:,d]
    p = c1 * c2 * np.exp(-c5/2)
    return p
```

```python
#test
x = np.array([[1,2],[2,1],[3,4]])
mu = np.array([1,2])
sigma = np.array([[1,0],[0,1]])
gauss(x,mu,sigma)
```

```python
X_range0 = [-3,3]
X_range1 = [-3,3]
```

```python
#显示等高线
def show_contour_gauss(mu,sig):
    xn = 40#等高线分辨率
    x0 = np.linspace(X_range0[0],X_range1[1],xn)
    x1 = np.linspace(X_range0[0],X_range1[1],xn)
    xx0, xx1 = np.meshgrid(x0,x1)
    x = np.c_[np.reshape(xx0,xn*xn,1),np.reshape(xx1,xn*xn,1)]
    f = gauss(x,mu,sig)
    f = f.reshape(xn,xn)
    f = f.T
    cont = plt.contour(xx0,xx1,f,15,color='k')
    plt.grid(True)
```

```python
#三维图形
def show3d_gauss(ax,mu,sig):
    xn = 40#等高线分辨率
    x0 = np.linspace(X_range0[0],X_range1[1],xn)
    x1 = np.linspace(X_range0[0],X_range1[1],xn)
    xx0, xx1 = np.meshgrid(x0,x1)
    x = np.c_[np.reshape(xx0,xn*xn,1),np.reshape(xx1,xn*xn,1)]
    f = gauss(x,mu,sig)
    f = f.reshape(xn,xn)
    f = f.T
    ax.plot_surface(xx0,xx1,f,rstride=2,cstride=2,alpha=0.3,color='blue',edgecolor='black')
```

```python
#主处理
mu = np.array([1,0.5])#图形中心
sigma = np.array([[2,1],[1,1]])#图形分布形状向上倾斜
Fig = plt.figure(1,figsize=(7,3))
Fig.add_subplot(1,2,1)
show_contour_gauss(mu,sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$',fontsize=14)
plt.ylabel('$x_1$',fontsize=14)
Ax = Fig.add_subplot(1,2,2,projection='3d')
show3d_gauss(Ax,mu,sigma)
Ax.set_zticks([0.05,0.10])
Ax.set_xlabel('$x_0$',fontsize=14)
Ax.set_ylabel('$x_1$',fontsize=14)
Ax.view_init(40,-100)
plt.show()
```
